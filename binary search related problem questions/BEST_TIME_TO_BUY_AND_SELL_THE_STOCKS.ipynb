```python
def findmaxprofit(price):
  maxprofit = 0
  minprice = float("inf")
  print(len(price))
  for i in range(len(price)):

    if price[i] < minprice:
      minprice = price[i]

    elif price[i]-minprice > maxprofit:
      maxprofit = price[i]-minprice
  return maxprofit



#driver code

days = [7,4,5,3,6,4,16]
result = findmaxprofit(days)

print("THE MAX PROFIT WE CAN ACHIEVE IS:",result)
```
