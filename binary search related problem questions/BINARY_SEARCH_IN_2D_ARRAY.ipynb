```python
def binarysearch(matrix, target):
  # Get the number of rows and columns from the matrix
  m = len(matrix)
  if m == 0:
    return False # Handle empty matrix case
  n = len(matrix[0])
  if n == 0:
    return False # Handle empty rows case

  left = 0
  right = (m * n) - 1

  while left <= right:
    mid = left + (right - left) // 2
    row = mid // n
    col = mid % n
    mid_element = matrix[mid//n][mid%n]

    if mid_element == target:
      return True
    elif mid_element < target:
      left = mid + 1
    else:
      right = mid - 1

  return False # Target not found

# Driver code
matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]]
target = 3
result = binarysearch(matrix, target)
print(f"Target {target} found: {result}")
```
